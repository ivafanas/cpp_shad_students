```c++
template<typename T>
void print(const T& x)
{
    std::cout << x << std::endl;
}

int main()
{
    print(5);
    print(6);
    print(5.0);
    print(6.0f);
    print("Dobrynia");
    print("Nikitich");
    print("Alesha");
    print("Popovich");
    print(std::string("Ilya"));
    print(std::string("Muromec"));
}
```

* Как быть с ODR?
* Плюсы и минусы шаблонов?
* Что в этом примере?

```c++
template<typename T>
const T& min(const T& a, const T& b)
{
    if (a < b)
        return a;
    else
        return b;
}

int main()
{
    const std::string s = min<std::string>("string 1", "string 2");
    std::cout << s << std::endl;
}
```

* Скомпилируется ли такой код?

```c++
template<typename T>
struct Vector3
{
    T x;
    T y;
    T z;
    
    Vector3<T> operator *(const float value)
    {
        return { x * value, y * value, z * value };
    }
};

std::string f()
{
    Vector3<std::string> v;
    return v.x;
}
```
