* сколько здесь аллокаций? какие?

```c++
std::deque<std::string> team;
for (int i = 0; i < N; ++i)
    team.emplace_back("imperial storm trooper");
```

* что будет выведено здесь?

```c++
std::vector<int> v = {10, 20, 30, 40, 50};

auto it1 = v.begin();
auto it2 = std::next(it1);
auto it3 = std::next(it2);
auto it4 = std::next(it3);

v.insert(it3, 25);

std::cout << *it1 << std::endl;
std::cout << *it2 << std::endl;
std::cout << *it3 << std::endl;
std::cout << *it4 << std::endl;
```

<details>
<summary>Подсказка</summary>
<p>

Все итераторы после вставки инвалидированы, разыменование инвалидированых итераторов - ub.

</p>
</details>
